```python
def get_avg(exp, runs, i):
    
    df_temp = pd.DataFrame()
    for run in runs:
        temp = pd.read_csv('results/final_results/' + exp + '/results_' + str(run) + str(i) + '.csv')
        df_temp = pd.concat((df_temp, temp))
    #     print(df_temp.shape)
    #     print(df_temp.columns)
    return df_temp

def plot(df_list):
    
    fig, ax = plt.subplots()
    for i, e in enumerate(df_list):
        temp = e
        plt.plot(range(len(temp)), temp)
    # ax.legend()
    plt.legend()
    plt.show()
```

```python
def creat_social_network(edges):

    G = nx.Graph()
    lines = edges.split("\n")
    for e in lines:
        nodes = e.split()
        if(len(nodes) > 1):
            if(nodes[0] not in G):
                G.add_node(nodes[0])
            if(nodes[1] not in G):
                G.add_node(nodes[1])

            G.add_edge(nodes[0], nodes[1])

    return G

def get_fb_network(df):

    fb_network = open('../data/facebook_combined.txt', 'r').read()
    G = creat_social_network(fb_network)
    A = nx.adjacency_matrix(G).todense()
    A = np.array(A)
    n = A.shape[0]
    G = nx.from_numpy_matrix(A)

    node_attr = df.set_index('id').to_dict('index')
    nx.set_node_attributes(G, node_attr)

    return G
```

```python
A = nx.adjacency_matrix(initial_graph).todense()
A = np.array(A)

def get_pol_diff(initial_data):
    
    pol_diff = []
    for n in initial_data['id']:
        n = int(n)
        neighbors = list(np.where(A[n] == 1)[0])
        temp_df = initial_data[initial_data['id'].isin(neighbors)]
        pol_inc_n = initial_data[initial_data['id'] == n]['new_pol_incl'].values[0]

        #print(temp_df.shape, len(neighbors), pol_inc_n)
        temp_df['pol_diff'] = temp_df['new_pol_incl'] - pol_inc_n
        std_pol_diff_n = (temp_df['pol_diff'] ** 2).sum()/temp_df.shape[0]
    #     sys.exit()
        pol_diff.append(std_pol_diff_n)
        
    return pol_diff
```

```python
def get_new_pol_incl(initial_data, a,b):
    
    new_pol_inc = []
    for i in range(initial_data.shape[0]):
        m = initial_data.iloc[i,a:b].mean()
        new_pol_inc.append(m)

    initial_data['new_pol_incl'] = new_pol_inc
    return initial_data
```
